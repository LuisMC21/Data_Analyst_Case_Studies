### Importing & merging many files

```python
import pandas as pd
```

```python
df_1930 = pd.read_csv("data/yob1930.txt", header = None, names = ["Name", "Gender", "Count"])
df_1930
```

```python
df_1931 = pd.read_csv("data/yob1931.txt", header = None, names = ["Name", "Gender", "Count"])
df_1931
```

```python
pd.concat(objs = [df_1930, df_1931], axis = 0, keys = [1930, 1931], names = ["Year"])
```

```python
pd.concat(objs = [df_1930, df_1931], axis = 0, keys = [1880, 1881],
          names = ["Year"]).droplevel(-1).reset_index()
```

```python
pd.read_csv("data/yob{}.txt".format(1930), header = None, names = ["Name", "Gender", "Count"])
```

```python
years = list(range(1923, 2019))
years
```

```python
dataframes = []
for year in years:
    data = pd.read_csv("data/yob{}.txt".format(year), header = None, 
                       names = ["Name", "Gender", "Count"])
    dataframes.append(data)
```

```python
dataframes
```

```python
len(dataframes)
```

```python
df = pd.concat(dataframes,  axis = 0, keys = years, names = ["Year"]).droplevel(-1).reset_index()
```

```python
df
```

```python
df.info()
```

```python
import seaborn as sns
```

## Most common names in a year

```python
df.sort_values('Count', ascending=False).groupby('Gender').head(1)
```

## TOP 5 most common names by gender

```python
df.groupby(['Name', 'Gender'])['Count'].sum().sort_values(ascending=False).reset_index().groupby('Gender').head(5)
```

## Evolution of the name James

```python
df[df['Name']=="James"].groupby(['Year', 'Name'])['Count'].sum().reset_index(1).plot()
```

## Evolution of the name Mary

```python
df[df['Name']=="Mary"].groupby(['Year', 'Name'])['Count'].sum().reset_index(1).plot()
```

## Less common names

```python
df.groupby(['Name', 'Gender'])['Count'].sum().sort_values(ascending=True).reset_index().groupby('Gender').head(5)
```

## names used for both sexes

```python
name_gender_counts = df.groupby('Name')['Gender'].nunique()

names_used_for_both_sexes = name_gender_counts[name_gender_counts > 1].index.tolist()

names_used_for_both_sexes
```

```python
df[df['Name'].isin(names_used_for_both_sexes)].groupby(['Name'])['Count'].sum().sort_values(ascending=False).reset_index().head(5)
```
